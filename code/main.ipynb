#### Train sales (Discounts data)

```python
discounts_data = sales_train.iloc[:, 7:]
discounts_data
```

```python
discounts_data.describe()
```

```python
# Calculate mean discount for each product
mean_discount = discounts_data.apply(lambda x: x.mean(), axis=1)
mean_discount
```

#### Encoding

```python
# Encode `warehouse` feature
sales_train.loc[:, 'warehouse'].unique()
```

```python
X = sales_train.copy()
X = X.drop(labels=['type_0_discount',
                   'type_1_discount',
                   'type_2_discount',
                   'type_3_discount',
                   'type_4_discount',
                   'type_5_discount',
                   'type_6_discount',
                   'availability'], axis=1)
X.head()
```

```python
X = X.dropna()
X.shape
```

```python
ordinal_encoder = OrdinalEncoder().fit(X.loc[:, ['warehouse']])
ordinal_encoder.categories_
```

```python
X['mean_discount'] = mean_discount.copy()
X['warehouse'] = ordinal_encoder.transform(X.loc[:, ['warehouse']])
X.head()
```

```python
# Select Out of Time (OOT) period
oot_date = '2023-11-01'
X_oot = X.loc[X.loc[:, 'date'] >= oot_date]
print(f'OOT percentage: {len(X_oot.index) / len(X.index):.4f}. Amount: {X_oot.shape[0]}')
```

```python
# Split data into train and test
X_train, X_test = train_test_split(X.loc[X.loc[:, 'date'] < oot_date], test_size=0.2, random_state=42)
print(f'Train shape: {X_train.shape}')
print(f'Test shape: {X_test.shape}')
```

```python
y_train = X_train.loc[:, 'sales']
y_test = X_test.loc[:, 'sales']
y_oot = X_oot.loc[:, 'sales']

X_train = X_train.drop(labels=['sales'], axis=1)
X_test = X_test.drop(labels=['sales'], axis=1)
X_oot = X_oot.drop(labels=['sales'], axis=1)
```

```python
# Use standardization for data
standard_scaler = StandardScaler().fit(X_train.loc[:, ['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']])
```

```python
standard_scaler.get_feature_names_out()
```

```python
X_train[['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']] = standard_scaler.transform(X_train.loc[:, ['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']])
X_test[['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']] = standard_scaler.transform(X_test.loc[:, ['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']])
X_oot[['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']] = standard_scaler.transform(X_oot.loc[:, ['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']])
```

```python
X_train.head()
```

#### Train Linear Regression

```python
features = ['warehouse', 'total_orders', 'sell_price_main', 'mean_discount']
```

```python
linear_regression = LinearRegression().fit(X_train.loc[:, features], y_train)
```

```python
# Check the performance
print(f'Train MAE: {mean_absolute_error(y_train, linear_regression.predict(X_train.loc[:, features]))}')
print(f'Test MAE: {mean_absolute_error(y_test, linear_regression.predict(X_test.loc[:, features]))}')
print(f'OOT MAE: {mean_absolute_error(y_oot, linear_regression.predict(X_oot.loc[:, features]))}')
```
